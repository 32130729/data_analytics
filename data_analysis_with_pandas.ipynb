```python
# Step 1: Import pandas
import pandas as pd

# Step 2: Load the CSV file into a DataFrame
df = pd.read_csv('/content/telecom_dataset.csv')

# Step 3: Inspect the DataFrame
print(df.head())  # View the first few rows
print(df.info())  # Summary of the DataFrame
print(df.shape)   # Shape of the DataFrame
```

```python
# Histogram
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('/content/telecom_dataset.csv')

# Creating a histogram using Seaborn
sns.histplot(data=df, x='customer service calls', bins=10, color='skyblue', kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of customer service calls')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creating a scatter plot using Matplotlib
plt.scatter(df['total day calls'], df['Predicted total day calls'], color='skyblue', alpha=0.5)
plt.xlabel('total day calls')
plt.ylabel('Predicted total day calls')
plt.title('total day calls vs Predicted total day calls')
plt.show()
```

```python
# Creating a box plot using Matplotlib
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['churn']==True]['customer service calls'], df[df['churn']==False]['customer service calls']], labels=['True', 'False'])
plt.xlabel('Churn')
plt.ylabel('customer service calls')
plt.title('customer service calls distribution by churn')
plt.show()
```

```python
# Creating a pair plot using Seaborn
sns.pairplot(df[['total day calls', 'total day minutes']], diag_kind='kde')
plt.show()
```

```python
calls_data = df[['total day calls', 'total eve calls', 'total night calls', 'total intl calls']]

# Calculate the correlation matrix
correlation_matrix = calls_data.corr()

# Plot the correlation matrix as a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
mean_total_day_calls = df['total day calls'].mean()

print(mean_total_day_calls)
```

```python
median_total_day_calls = df['total day calls'].median()

print(median_total_day_calls)
```

```python
std_total_day_calls = df['total day calls'].std()

print(std_total_day_calls)
```

```python
max_total_day_calls = df['total day calls'].max()

print(max_total_day_calls)

min_total_day_calls = df['total day calls'].min()

print(min_total_day_calls)
```
